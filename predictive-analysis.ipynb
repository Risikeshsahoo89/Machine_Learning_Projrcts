```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/cardataset/data.csv")
```

```python
df
```

```python
df.info()
```

```python
df['Make'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Make_encoded'] = le.fit_transform(df['Make'])
```

```python
subset_df = df[['Make','Make_encoded']].drop_duplicates()
```

```python
print(subset_df)
```

```python
df.head()
```

```python
df['highway MPG']=df['highway MPG'].astype(int)
```

```python
df.dropna(subset=['Engine HP'], inplace=True)
```

```python
df['Engine HP']=df['Engine HP'].astype(int)
```

```python
df.dropna(subset=['Engine Fuel Type'], inplace=True)
```

```python
df['Engine Fuel Type'].unique()
```

```python
le = LabelEncoder()
df['Engine Fuel Type_encoded'] = le.fit_transform(df['Engine Fuel Type'])
```

```python
subset_df = df[['Engine Fuel Type','Engine Fuel Type_encoded']].drop_duplicates()
print(subset_df)
```

```python
df['Transmission Type'].unique()
```

```python
df['Transmission Type']=df['Transmission Type'].replace({'MANUAL':0, 'AUTOMATIC':1,'AUTOMATED_MANUAL':2, 'DIRECT_DRIVE':3, 'UNKNOWN':4})
```

'MANUAL':0, 'AUTOMATIC':1,'AUTOMATED_MANUAL':2, 'DIRECT_DRIVE':3, 'UNKNOWN':4

```python
df['Driven_Wheels'].unique()
```

```python
df['Driven_Wheels']=df['Driven_Wheels'].replace({'rear wheel drive':0, 'front wheel drive':1,'all wheel drive':2, 'four wheel drive':3})
```

'rear wheel drive':0, 'front wheel drive':1,'all wheel drive':2, 'four wheel drive':3

```python
print(df['Number of Doors'].unique())
```

```python
df.dropna(subset=['Number of Doors'], inplace=True)
```

```python
print(df['Number of Doors'].unique())
```

```python
df['Number of Doors']=df['Number of Doors'].astype(int)
```

```python
df['Engine Cylinders'].unique()
```

```python
df.dropna(subset=['Engine Cylinders'], inplace=True)
```

```python
df['Engine Cylinders']=df['Engine Cylinders'].astype(int)
```

```python
df['Vehicle Size'].unique()
```

```python
df['Vehicle Size']=df['Vehicle Size'].replace({'Compact':0, 'Midsize':1,'Large':2})
```

'Compact':0, 'Midsize':1,'Large':2

```python
x = df[['Make_encoded', 'Engine Fuel Type_encoded', 'Transmission Type', 'Driven_Wheels', 'Number of Doors', 'Engine Cylinders', 'Vehicle Size', 'highway MPG','Engine HP']]
```

```python
x
```

```python
y = df[['MSRP']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
print("r2_score", r2_score*100)
```
